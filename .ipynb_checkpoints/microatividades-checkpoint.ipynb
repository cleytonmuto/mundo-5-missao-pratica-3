**importar a biblioteca pandas**

```python
import pandas as pd
```

**criar a variável para receber o conteúdo do arquivo CSV (picoweb.csv), passando como parâmetros o separador de colunas (;), a engine - com o valor 'python' e o enconding (UTF-8) relativo aos dados constantes no arquivo lido (esse último parâmetro pode ser opcional, dependendo do enconding existente);atribuir os dados lidos do CSV à variável criada anteriormente; salvar as alterações; imprimir/exibir em tela os dados da variável.**

```python
data = pd.read_csv('picoweb.csv', sep=';', engine='python', encoding='utf-8')
```

**no mesmo arquivo/script utilizado na microatividade 1, criar uma nova variável (subset); atribuir, a essa nova variável (subset), um subconjunto de dados contendo apenas parte das colunas (recomenda-se a utilização de 3 colunas) disponíveis no conjunto de dados original;**

```python
subset = data[['ID','Duration','Pulse']]
display(subset)
```

**usando as opções de configuração da biblioteca pandas, defina um novo valor para a propriedade “max_rows”, definindo o novo valor para 9999;**

```python
pd.set_option('display.max_rows',9999)
display(data)
```
